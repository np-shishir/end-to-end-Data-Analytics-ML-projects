```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('superstore.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.shape
```

```python
ip = df.iloc[:, [4, 7, 9,15, 17, 19]]
```

```python
ip
```

**One Hot Encoding to convert text columns into numeric**

```python
input = pd.get_dummies(ip, drop_first=True)
```

```python
output = df.iloc[:, [20]]
```

```python
output
```

```python
from sklearn.model_selection import train_test_split
```

```python
i_train, i_test, o_train, o_test = train_test_split(input, output, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

**Linear Regression Model**

```python
model = LinearRegression()
```

```python
model.fit(i_train, o_train)
```

```python
o_predict = model.predict(i_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
mae = mean_absolute_error(o_test, o_predict)
```

```python
r2 = r2_score(o_test, o_predict)
```

```python
print(f"Mean Absolute error: {mae}")
print(f"R2 Score: {r2}")
```

**Random Forest Model**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
rf_model.fit(i_train, o_train)
```

```python
rf_o_predict = rf_model.predict(i_test)
```

```python
mae_rf = mean_absolute_error(o_test, rf_o_predict)
r2_rf = r2_score(o_test, rf_o_predict)
```

```python
print(f"Random Forest MAE: {mae_rf}")
print(f"Random Forest R2 Score: {r2_rf}")
```

**Random Forest model was way better than Linear Regression model**
